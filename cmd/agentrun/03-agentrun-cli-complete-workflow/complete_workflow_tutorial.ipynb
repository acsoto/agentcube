## Step 4: Packaging Your Agent with `agentrun pack`

The first step in the AgentRun CLI workflow is to package your agent. This command:
- Validates your agent structure
- Creates an `agent_metadata.yaml` configuration file
- Generates a Dockerfile for containerization
- Prepares the agent workspace for building

### 4.1 Understanding Pack Options

Let's first see what options are available:

````python
!agentrun pack --help
````

### 4.2 Packaging the Agent

Now let's package our agent:

````python
!agentrun pack \
    -f ./agent \
    --agent-name "conversational-assistant" \
    --description "A conversational agent with tool-calling capabilities" \
    --language python \
    --entrypoint "python main.py" \
    --port 8080 \
    --build-mode local \
    --verbose
````

### 4.3 Examining the Generated Files

Let's examine the files that were generated:

#### Agent Metadata

````python
# View the generated agent_metadata.yaml
!cat agent/agent_metadata.yaml
````

#### Dockerfile

````python
# View the generated Dockerfile
!cat agent/Dockerfile
````

### 4.4 Understanding the Metadata

The `agent_metadata.yaml` file contains important configuration:

- **agent_name**: Unique identifier for your agent
- **description**: Human-readable description
- **language**: Programming language (python, java)
- **entrypoint**: Command to start the agent
- **port**: Port the agent listens on
- **build_mode**: local or cloud
- **requirements_file**: Dependencies file

This metadata drives the entire deployment process.

### 6.2 Publishing the Agent

For **local build mode**, you need to specify the image URL and credentials:

````python
# Publish the agent to AgentCube
# Uncomment and modify the following command with your registry details:

# !agentrun publish \
#     -f ./agent \
#     --version "v1.0.0" \
#     --image-url "docker.io/your-username/conversational-assistant" \
#     --image-username "your-username" \
#     --image-password "your-password" \
#     --description "Conversational assistant with tool calling" \
#     --verbose

print("Publish command template:")
print("")
print("agentrun publish \\")
print("    -f ./agent \\")
print("    --version 'v1.0.0' \\")
print("    --image-url 'docker.io/username/conversational-assistant' \\")
print("    --image-username 'your-username' \\")
print("    --image-password 'your-password' \\")
print("    --verbose")
print("")
print("Note: This step requires:")
print("  1. A container registry account (Docker Hub, ECR, GCR, etc.)")
print("  2. AgentCube API endpoint configured")
print("  3. Proper authentication credentials")
````

### 6.3 What Happens During Publishing?

When you run `agentrun publish`, the following happens:

1. **Image Tagging**: The local image is tagged with version (e.g., `v1.0.0`)
2. **Registry Login**: Authenticates with the container registry
3. **Image Push**: Pushes the image to the registry
4. **AgentCube Registration**: Calls AgentCube API to register the agent
5. **Metadata Update**: Updates `agent_metadata.yaml` with:
   - agent_id
   - agent_endpoint
   - version
   - image URL

## Step 9: Complete Workflow Summary

Let's summarize the complete AgentRun CLI workflow:

### Workflow Steps

```bash
# 1. Package the agent
agentrun pack -f ./agent --agent-name "my-agent" --verbose

# 2. Build container image
agentrun build -f ./agent --verbose

# 3. Publish to AgentCube
agentrun publish -f ./agent \
    --version "v1.0.0" \
    --image-url "docker.io/username/my-agent" \
    --image-username "username" \
    --image-password "password" \
    --verbose

# 4. Check status
agentrun status -f ./agent --verbose

# 5. Invoke the agent
agentrun invoke -f ./agent --payload '{"prompt": "Hello!"}' --verbose
```

### Key Benefits

1. **Standardization**: Consistent packaging and deployment process
2. **Simplicity**: Single command for each step
3. **Flexibility**: Support for local and cloud builds
4. **Traceability**: Metadata tracking throughout the lifecycle
5. **Integration**: Python SDK for CI/CD integration

## Step 10: Advanced Topics

### 10.1 Using the Python SDK

AgentRun CLI also provides a Python SDK for programmatic access:

````python
# Example: Using the Python SDK
from agentrun.runtime import pack_runtime, build_runtime, publish_runtime

# This allows you to integrate AgentRun into your Python applications
# and CI/CD pipelines

print("Python SDK Example:")
print("")
print("from agentrun.runtime import pack_runtime, build_runtime, publish_runtime")
print("")
print("# Pack agent")
print("pack_runtime.pack(workspace='./agent', agent_name='my-agent')")
print("")
print("# Build image")
print("build_runtime.build(workspace='./agent')")
print("")
print("# Publish to AgentCube")
print("publish_runtime.publish(workspace='./agent', version='v1.0.0')")
````

### 10.2 CI/CD Integration

Example GitHub Actions workflow:

```yaml
name: Deploy Agent

on:
  push:
    branches: [main]

jobs:
  deploy:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v2
      
      - name: Install AgentRun CLI
        run: pip install agentrun
      
      - name: Pack Agent
        run: agentrun pack -f ./agent --agent-name "my-agent"
      
      - name: Build Image
        run: agentrun build -f ./agent
      
      - name: Publish to AgentCube
        run: |
          agentrun publish -f ./agent \
            --version "${{ github.sha }}" \
            --image-url "${{ secrets.REGISTRY_URL }}/my-agent" \
            --image-username "${{ secrets.REGISTRY_USERNAME }}" \
            --image-password "${{ secrets.REGISTRY_PASSWORD }}"
```

## Step 11: Troubleshooting

### Common Issues

#### 1. "Docker is not available"
**Solution**: Install Docker and ensure it's running
```bash
docker --version
docker ps
```

#### 2. "Metadata file not found"
**Solution**: Run `agentrun pack` first
```bash
agentrun pack -f ./agent
```

#### 3. "Agent not published yet"
**Solution**: Complete the build and publish steps
```bash
agentrun build -f ./agent
agentrun publish -f ./agent --version v1.0.0
```

#### 4. "Invalid requirements.txt"
**Solution**: Ensure requirements.txt exists and is properly formatted

#### 5. "Build failed"
**Solution**: Check Docker logs and Dockerfile syntax
```bash
agentrun build -f ./agent --verbose
```
